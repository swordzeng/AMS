```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('AMS.db')
query = """select H.Date, H.AccountID, H.SymbolCode, H.SymbolName, H.CurTrade, H.Qty, H.Cost, C.Close, E.Close as ExRate
        from Holding_Table_Test as H left join Close_Price as C on H.Date = C.Date and H.SymbolCode = C.SymbolCode
        left join Exchange_Rate as E on H.Date = E.Date and H.CurTrade = E.SymbolCode
        where H.Date = '2019-12-31'"""
df = pd.read_sql(query, con=conn)
df
```

```python
holding = df
holding.insert(5,'QtyBuy',0)
holding.insert(6,'CostBuy',0)
holding.insert(7,'QtySell',0)
holding.insert(8,'PriceSell',0)
holding.insert(12,'MV',0)
holding.insert(13,'PL_Realized',0)
holding.insert(14,'PL_Float',0)
holding.insert(15,'PL_Interest',0)
holding
```

```python
holding['Close'] = holding.apply(lambda x: 1 if x['SymbolCode'] in ['CNY','HKD','USD'] else x['Close'],axis=1)
holding['MV'] = holding.apply(lambda x: x['Qty']*x['Close'],axis=1)
holding['PL_Float'] = holding.apply(lambda x: x['Qty']*(x['Close']-x['Cost']),axis=1)
holding['ExRate'] = holding.apply(lambda x: 1 if x['CurTrade'] in ['CNY'] else x['ExRate'],axis=1)
holding
```

```python
holding.to_sql('Holding_Table', con=conn, if_exists='append', index=False)
```

```python
cursor = conn.cursor()
query = "update Holding_Table set PL_Float = round(PL_Float,2)"
cursor.execute(query)
conn.commit()
```

```python
query = "update Holding_Table set MV = round(MV,2)"
cursor.execute(query)
conn.commit()
```
